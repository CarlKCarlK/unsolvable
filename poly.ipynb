```python
import sympy as sym
from sympy.abc import x

equation = x**5 - x - 1
solution = sym.solve(equation, x)
print(solution)
```

```python
from sympy import N, CRootOf

print(N(CRootOf(x**5 - x - 1, 0)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 1
x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

plt.figure(figsize=(6,4))
plt.axhline(0, color='black',linewidth=0.5)  # x-axis
plt.axvline(0, color='black',linewidth=0.5)  # y-axis
plt.title('Plot of $x^2 + 1$')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x_vals, y_vals, label=r'$x^2 + 1 = 0$')
plt.show()
```

```python
from sympy import solve, Eq, CRootOf, N, degree
from sympy.abc import x

equation = Eq(x**2 + 1, 0)

numerical_solution = [N(CRootOf(equation, d)) for d in range(degree(equation))]
print(numerical_solution)
```
